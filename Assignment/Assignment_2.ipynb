1. Email Validator (String Filtering + Loops)

```python
email = input("Enter your email: ")

if "@" in email and email.endswith(".com") or email.endswith(".org"):
    print("Valid Email")

else:
    print("Invalid Email")   
```

2. Count Word Frequency in Customer Reviews

```python
review = "This product is good and this service is excellent"

review = review.lower()
word = review.split()
word_count = {}

for i in word:
    if i in word_count:
        word_count [i]+= 1

    else:
        word_count[i] = 1   


print(word_count)        

    
```

3. Password Strength Checker

```python
Password =input("Enter your Password: ")


upper = False
lower = False
digit = False
special = False

for ch in Password:
    if ch.isupper():
        upper = True

    elif ch.islower():
        lower = True

    elif ch.isdigit():
        digit = True 

    elif not ch.isalnum():
        special= True  

if upper and lower and digit and special:
    print("Strong Password") 

else:
    print("Weak Password")              

       
```

5. SMS Word Counter

```python
message =input("Enter your message: ")

mess  = message.split()
Words_count =len(mess)

Characters_count = len(message)

if Characters_count < 160 :
    print("Status = Within Limit")

else:
    print("Status =Not Within Limit")   

print("Words = ",Words_count,"Characters = ",Characters_count)
```

6. Remove Duplicates from Contact List

```python
names = input (" Enter your names: ")

names = names.title()
name_list= names.split()
list = []

for name in name_list:
        
        if  name not in list:
                list.append(name)
            
print(",".join(list))        

    
```

7. Detect Suspicious Words in Text (Moderation)

```python

banned_words = ['spam', 'scam', 'fake']
text =input("Enter a text :")

text = text.lower()
for word in banned_words:
    if word in text:
        print("Warning: Suspicious content detected")

        break  
```

8. Invoice Code Formatter

```python
num =input(" Enter an invoice number: ")

number =len(num)

zero_need = 5 - number
zero = "0" * zero_need

invoice_number = "INV"+ zero + num


print(invoice_number)
```

9. Count Hashtags and Mentions (Social Media Analytics)

```python
post =input("Enter your message: ")

Hashtags_count = 0
Mentions_count = 0
for mess in post:
    if mess == "#":
        Hashtags_count += 1

    elif mess == "@":
        Mentions_count += 1 

print("Hashtags=", Hashtags_count,"Mentions=",Mentions_count)      

     
```

10. Word Reversal for Data Encryption

```python
word =input(" Enter a sentence: ")
words= word.split()
Reverse_word =  []

for w in words:
        Reverse_word.append(w[::-1])
      
print(" ".join(Reverse_word)) 
```
